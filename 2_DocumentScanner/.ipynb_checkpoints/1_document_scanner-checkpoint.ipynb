```python
img_orig = cv2.imread('./images/03.jpg')

cv2.namedWindow('for_test', cv2.WINDOW_NORMAL)
cv2.imshow('for_test', img_orig)
cv2.waitKey()
cv2.destroyAllWindows()
```

### Scanning

```python
# 保持比例調整大小
```

```python
def resizer(image, width=500) :
    # 建立 width 和 height
    h, w, c = image.shape
    
    height = int((h/w) * width)
    size = (width, height)
    image = cv2.resize(image, (width, height))
    return image, size
```

```python
# 測試
img_re, size = resizer(img_orig)

print(size)
cv2.imshow('Resizer', img_re)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 細節增強 , 將名片邊緣顯現
# sigma_s：這是控制空間濾波器的標準差（Standard Deviation）的參數。它用於平滑影像的細節。較小的 sigma_s 值會導致較強的平滑效果，而較大的值則會導致較弱的平滑效果。換句話說，它影響了平滑的程度。

# sigma_r：這是控制強度增強濾波器的標準差的參數。它用於增強影像的細節。較小的 sigma_r 值會導致較強的細節增強，而較大的值則會導致較弱的細節增強。換句話說，它影響了細節增強的程度。
```

```python
detail = cv2.detailEnhance(img_re, sigma_s=20, sigma_r= 0.15)

gray = cv2.cvtColor(detail, cv2.COLOR_BGR2GRAY)    # 轉為灰階

blur = cv2.GaussianBlur(gray, (5, 5), 0)   # 減少圖像中的噪點、平滑細節或降低圖像的高頻細節信息

# 邊緣檢測
edge_image = cv2.Canny(blur, 75, 200)

# morphological transform
kernel = np.ones((5, 5), np.uint8)   # 建立 5X5 的核心
dilate = cv2.dilate(edge_image, kernel, iterations=1)   # 膨脹 , 用意是擴大或加強圖像中的白色區域（也稱為前景），並將它們連接在一起
closing = cv2.morphologyEx(dilate, cv2.MORPH_CLOSE, kernel)    # 用於去除二值圖像中的小孔洞，並連接物體的細小間隙

# 檢測輪廓
contours, hire = cv2.findContours(closing, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
contours = sorted(contours, key=cv2.contourArea, reverse=True)
for contour in contours:
    peri = cv2.arcLength(contour, True)   # 第一個參數是輪廓，第二個參數是布爾值（True 或 False），用於指定輪廓是否封閉。計算周長是為了後續的多邊形近似
    approx = cv2.approxPolyDP(contour, 0.02*peri, True)    # 將輪廓近似為具有較少頂點的多邊形，以簡化輪廓的表示

    if len(approx) == 4 :                  # 如果滿足四邊形的輪廓
        four_points = np.squeeze(approx)
        break

cv2.drawContours(img_re, [four_points], -1, (0,255,0), 3)

# 找原圖的四角形座標
multiplier = img_orig.shape[1] / size[0]
four_points_orig = four_points * multiplier
four_points_orig = four_points_orig.astype(int)
```

```python
wrap_img = four_point_transform(img_orig, four_points_orig)
```

### 建立 document scanner function

```python
def document_scanner(image, width=500) :
    img_re, size = resizer(image)
    
    detail = cv2.detailEnhance(img_re, sigma_s=20, sigma_r= 0.15)

    gray = cv2.cvtColor(detail, cv2.COLOR_BGR2GRAY)    # 轉為灰階
    
    blur = cv2.GaussianBlur(gray, (5, 5), 0)   # 減少圖像中的噪點、平滑細節或降低圖像的高頻細節信息
    
    # 邊緣檢測
    edge_image = cv2.Canny(blur, 75, 200)
    
    # morphological transform
    kernel = np.ones((5, 5), np.uint8)   # 建立 5X5 的核心
    dilate = cv2.dilate(edge_image, kernel, iterations=1)   # 膨脹 , 用意是擴大或加強圖像中的白色區域（也稱為前景），並將它們連接在一起
    closing = cv2.morphologyEx(dilate, cv2.MORPH_CLOSE, kernel)    # 用於去除二值圖像中的小孔洞，並連接物體的細小間隙
    
    # 檢測輪廓
    contours, hire = cv2.findContours(closing, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
    contours = sorted(contours, key=cv2.contourArea, reverse=True)
    for contour in contours:
        peri = cv2.arcLength(contour, True)   # 第一個參數是輪廓，第二個參數是布爾值（True 或 False），用於指定輪廓是否封閉。計算周長是為了後續的多邊形近似
        approx = cv2.approxPolyDP(contour, 0.02*peri, True)    # 將輪廓近似為具有較少頂點的多邊形，以簡化輪廓的表示
    
        if len(approx) == 4 :                  # 如果滿足四邊形的輪廓
            four_points = np.squeeze(approx)
            break
    
    cv2.drawContours(img_re, [four_points], -1, (0,255,0), 3)
    
    # 找原圖的四角形座標
    multiplier = image.shape[1] / size[0]
    four_points_orig = four_points * multiplier
    four_points_orig = four_points_orig.astype(int)

    wrap_img = four_point_transform(image, four_points_orig)

    return wrap_img, four_points_orig, img_re, closing
```

```python
img = cv2.imread('./images/001.jpg')
wrapimg, points, cnt_img, edgeimg = document_scanner(img)
```

```python
cv2.imshow('original', img)
cv2.imshow('resize', cnt_img)
cv2.imshow('wrap', wrapimg)
cv2.imshow('edge', edgeimg)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 調整亮度和對比
```

```python
def apply_brightness_contrast(input_img, brightness = 0, contrast = 0):
    
    if brightness != 0:
        if brightness > 0:
            shadow = brightness
            highlight = 255
        else:
            shadow = 0
            highlight = 255 + brightness
        alpha_b = (highlight - shadow)/255
        gamma_b = shadow
        
        buf = cv2.addWeighted(input_img, alpha_b, input_img, 0, gamma_b)
    else:
        buf = input_img.copy()
    
    if contrast != 0:
        f = 131*(contrast + 127)/(127*(131-contrast))
        alpha_c = f
        gamma_c = 127*(1-f)
        
        buf = cv2.addWeighted(buf, alpha_c, buf, 0, gamma_c)

    return buf
```

```python
rgb = cv2.cvtColor(wrapimg, cv2.COLOR_BGR2RGB)
magic_image = apply_brightness_contrast(wrapimg, 40, 0)
magic_image_c1 = apply_brightness_contrast(wrapimg, 120, 0)
```

```python
cv2.imshow('magic', magic_image)
cv2.imshow('wrap', wrapimg)
cv2.waitKey()
cv2.destroyAllWindows()
```
