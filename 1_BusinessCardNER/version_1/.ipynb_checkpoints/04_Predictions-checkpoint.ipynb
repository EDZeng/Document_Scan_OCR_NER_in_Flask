```python
import numpy as np
import pandas as pd
import cv2
import pytesseract
from glob import glob
import spacy
import re
import string
```

```python
def cleanText(txt):
    punctuation = '!"#$%&\'()*+:;<=>?[\\]^`{|}~'  # 有些需要保留像是@和其他東西
    whitespace = string.whitespace
    tableWhitespace = str.maketrans('','',whitespace)
    tablePunctuation = str.maketrans('','',punctuation)
    text = str(txt)
    text = text.lower()
    removewhitespace = text.translate(tableWhitespace)
    removepunctuation = removewhitespace.translate(tablePunctuation)
    
    return str(removepunctuation)
```

### 載入 NER model

```python
model_ner = spacy.load('./output/model-best/')
```

```python
# 載入圖片
image = cv2.imread('./data/6.jpg')
# cv2.imshow('ofr_test', image)
# cv2.waitKey()
# cv2.destroyAllWindows()

# 用 pytesseract 提取數據
tessData = pytesseract.image_to_data(image)
# 轉換成表格並做清理
tessList = list(map(lambda x : x.split('\t'), tessData.split('\n')))
df = pd.DataFrame(tessList[1:], columns=tessList[0])
df.dropna(inplace=True)
df['text'] = df['text'].apply(cleanText)

# 將資料轉換成內容
df_clean = df.query("text != '' ")
content = " ".join([w for w in df_clean['text']])   # 將所有內容拼在一起再用空格隔開
# print(content)
# 放入模型預測
doc = model_ner(content)


```

### 將個別被拆開的實體變成一個完整的實體

```python
# 先將 label 前面的分類移除
bb_df['label'] = bb_df['label'].apply(lambda x : x[2:])
bb_df
```

```python
# 分組標籤
class groupgen():
    def __init__(self):
        self.id = 0
        self.text = ''

    def getgroup(self, text):          # 舉例對照 PHONE 是不是等於 PHONE , 是的話就是歸類在同一類
        if self.text == text :
            return self.id
        else :
            self.id += 1
            self.text = text
            return self.id

grp_gen = groupgen()
```

```python
bb_df['group'] = bb_df['label'].apply(grp_gen.getgroup)
```

```python
bb_df    # 這邊就可以看到group欄位被歸類在同一個欄位
```

```python
img_tagging = group_tag_img.agg({
    'left' : min,
    'right' : max,
    'top' : min,
    'bottom' : max,
    'label' : np.unique,
    'tokens' : lambda x : " ".join(x),
})
```

### 解析

```python
def parser(text,label) :
    if label  == 'PHONE':
        text = text.lower()
        text = re.sub(r'\D', '', text)

    elif label == "EMAIL":
        text = text.lower()
        allow_spcial_chart = '@_.\-'
        text = re.sub(r'[^A-Za-z0-9{} ]'.format(allow_spcial_chart), '', text)

    elif label == 'WEB':
        text = text.lower()
        allow_spcial_chart = ':/%#\-'
        text = re.sub(r'[^A-Za-z0-9{} ]'.format(allow_spcial_chart), '', text)

    elif label in ('NAME', 'DES'):
        text = text.lower()
        text = re.sub(r'[^a-z ]', '', text)
        text = text.title()

    elif label == 'ORG' :
        text = text.lower()
        text = re.sub(r'[^a-z0-9 ]', '', text)
        text = text.title()

    return text
```

### 實體

```python
info_array = dataframe_info[['tokens', 'label']].values
entities = dict(NAME=[], ORG=[], DES=[], PHONE=[], EMAIL=[], WEB=[])

previous = 'O'

for tokens, label in info_array:
    # print(token, label)
    bio_tag = label[0:1]
    label_tag = label[2:]
    # print(bio_tag, label_tag)

    # 步驟一 , 傳遞tokens
    text = parser(tokens, label_tag)      # 傳遞 tokens 內容和經過處理的 label 標籤到 parser function


    if bio_tag in ('B', 'I') :
        if previous != label_tag :
            entities[label_tag].append(text)
        else :
            if bio_tag == "B" :
                entities[label_tag].append(text)
            else :
                if label_tag in ('NAME', 'DES', 'ORG') :
                    entities[label_tag][-1] = entities[label_tag][-1] + " " + text
                else :
                    entities[label_tag][-1] = entities[label_tag][-1] + text
        
    previous = label_tag
```
